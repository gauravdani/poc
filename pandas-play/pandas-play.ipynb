```python
plt.plot(df['DATE'], df['TMAX'], linestyle='--', color='red', linewidth=2)

plt.title('Customized Line Plot')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')

plt.show()
```

```python
plt.plot(df['DATE'], df['TMIN'], linestyle='--', color='red', linewidth=2)

plt.title('Customized Line Plot')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')

plt.show()
```

```python
df['7-Day Moving Average'] = df['TMAX'].rolling(window=7).mean()
```

```python
plt.plot(df['DATE'], df['7-Day Moving Average'], label='7-Day Moving Average', color='red', linewidth=2)
```

```python
df['7-Day Moving Average'] = df['TMIN'].rolling(window=7).mean()
plt.plot(df['DATE'], df['TMIN'], label='Daily Temperature')
plt.plot(df['DATE'], df['7-Day Moving Average'], label='7-Day Moving Average', color='red', linewidth=2)

plt.title('Daily Temperature with 7-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()

plt.show()
```
